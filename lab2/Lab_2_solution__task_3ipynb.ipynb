```python
# Install Ultralytics YOLOv8 (if not installed)
!pip install ultralytics

# Import necessary libraries
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os

# Load the pre-trained YOLOv8 model
model = YOLO('yolov8n.pt')  # Using the small YOLOv8 model for efficiency

# Define the path to the image
image_path = '/content/drive/My Drive/1.png'


# Perform object detection
results = model(image_path)

# Display the detected image
result_image = results[0].plot()  # Plot image with detections
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
